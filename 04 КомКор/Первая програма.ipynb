```python
#Переобразовываем метод object в float столбец цена
def price(s):
    i = str(s).find(',')
    if(i>0):
        return s[:i] + '.' + s[i+1:]
    else :
        return s 
    
def tochka(s):
    i = str(s).find('.')
    if(i>0):
        return s[:i] + '' + s[i+1:]
    else :
        return s 
```

```python
belu['Цена'] = belu['Цена'].apply(tochka)
gche['Цена'] = gche['Цена'].apply(tochka)
mgnt['Цена'] = mgnt['Цена'].apply(tochka)
mtss['Цена'] = mtss['Цена'].apply(tochka)
nvtk['Цена'] = nvtk['Цена'].apply(tochka)
phor['Цена'] = phor['Цена'].apply(tochka)
pikk['Цена'] = pikk['Цена'].apply(tochka)
```
