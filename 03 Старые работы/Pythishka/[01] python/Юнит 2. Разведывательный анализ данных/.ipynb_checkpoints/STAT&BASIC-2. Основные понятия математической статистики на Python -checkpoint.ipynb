# 2.8. Как найти вероятность X через Z?

```python
from scipy.stats import norm
```

```python
ror = norm.cdf(1)
round(ror, 2)
```

```python
ror = norm.cdf(1.12)
round(ror, 2)
```

```python
ror = norm.cdf(-1.87)
round(ror, 2)
```

```python
ror = norm.ppf(0.1)
round(ror, 2)
```

```python
ror = norm.ppf(0.95)
round(ror, 2)
```

```python
ror = norm.ppf(0.025)
round(ror, 2)
```

```python
ror = norm.ppf(0.005)
round(ror, 2)
```

```python
ror = norm.ppf(0.99)
round(ror, 2)
```
