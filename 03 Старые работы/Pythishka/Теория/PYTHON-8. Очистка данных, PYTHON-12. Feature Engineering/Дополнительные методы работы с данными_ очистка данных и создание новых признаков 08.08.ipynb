## Заполняем пустые значения о возрасте

```python
# Способ 1: Среднее арифметическое
titanic['Age'].fillna(titanic.Age.mean(), inplace = True)
```

```python
# Способ 2: Cредние арифметические в зависимости от класса каюты
age_1 = titanic[titanic['Pclass'] == 1]['Age'].median()
print(age_1)
age_2 = titanic[titanic['Pclass'] == 2]['Age'].median()
print(age_2)
age_3 = titanic[titanic['Pclass'] == 3]['Age'].median()
print(age_3)
```

```python
#Функция fill_age()
def fill_age(row):
    if pd.isnull(row['Age']):
        if row['Pclass'] == 1:
            return age_1
        if row['Pclass'] == 2:
            return age_2
        if row['Pclass'] == 3:
            return age_3
    return row['Age']
```

```python
titanic['Age'] = titanic.apply(fill_age, axis = 1)
titanic
```

```python
titanic.info()
```


### Создаем новый столбец с информацией о том, был ли пассажир на борту один или с родственниками
### Feature Engineering 

```python
# Способ 1: с помощью именной функции и apply
def is_alone(row):
    if row['SibSp'] + row['Parch'] == 0:
        return 'был один'
    return 'был не один'
```

```python
titanic['Is_Alone'] = titanic.apply(is_alone, axis = 1)
```

```python
titanic
```

```python
# Способ 2: с помощью lambda-функции
titanic['Alone'] = titanic.apply(lambda x: 'один' if x['SibSp'] + x['Parch'] == 0 else 'не один', axis = 1)
```

```python
titanic.drop('Is_Alone', axis = 1, inplace = True)
```

```python
# Построим сводную таблицу (pivot_table), в которой отразим количество погибших и выживших 
# из числа путешествовавших в одиночку или с родственниками
titanic.pivot_table(columns='Alone', index='Survived', values='Age',aggfunc='count')
```

```python
titanic.pivot_table(columns='Alone', index='Survived', aggfunc='count')
```
